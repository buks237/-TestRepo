```python
# Task 2: Mark the success/failed launches for each site on the map

import folium

# Create a base map centered on average coordinates
launch_map = folium.Map(location=[spacex_df["Lat"].mean(), spacex_df["Long"].mean()], zoom_start=5)

# Replace 'class' below with the actual name of the success/failure column
for idx, row in spacex_df.iterrows():
    marker_color = "green" if row["class"] == 1 else "red"  # adjust column name if needed

    folium.Marker(
        location=[row["Lat"], row["Long"]],
        popup=f"""Index: {idx}<br>
                  Launch Site: {row['Launch Site']}<br>
                  Outcome: {'Success' if row['class'] == 1 else 'Failure'}""",
        icon=folium.Icon(color=marker_color)
    ).add_to(launch_map)

launch_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import radians, sin, cos, sqrt, atan2

def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in kilometers

    # Convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance
orlando_lat = 28.5383
orlando_lon = -81.3792

# Create a new column for distance to Orlando
spacex_df['distance_to_orlando_km'] = spacex_df.apply(
    lambda row: calculate_distance(row['Lat'], row['Long'], orlando_lat, orlando_lon),
    axis=1
)

# View result
spacex_df[['Launch Site', 'Lat', 'Long', 'distance_to_orlando_km']].head()
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
